```python
import numpy as np
import matplotlib.pyplot as plt

# X ve Y verilerini tanımlıyoruz
x = np.array([8, 7, 5, 7, 6 ,5,4,2,2,1,2,2,0,0,-1,0,0,-1])
y = np.array([0, 50, 50, 50, 100,50,0,0,100,80,100,0,0,100,80,100,0,0])

# Grafik çizimi
plt.plot(x, y)

# Başlık
plt.title("11-A Sınıfı Grafik Çizimi")

# Yatay ve dikey eksen çizgilerini ekliyoruz
plt.axhline(0, color='black',linewidth=1)  # Yatay eksen (X=0)
plt.axvline(0, color='black',linewidth=1)  # Dikey eksen (Y=0)

# X ve Y eksenleri için etiketler ekliyoruz
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")

# Grafik gösterimi
plt.show()
```
